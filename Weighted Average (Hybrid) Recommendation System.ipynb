```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
movies.head()
```

```python
##We can see that id in movies is simlar to movie_id in credits, therefore to merge both dataset we will use id column
```

```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id": "id"})
```

```python
credits.head()
```

```python
credits_column_renamed.head()
```

```python
##movie_id renamed to id
```

```python
movies_merged = movies.merge(credits_column_renamed, on='id')
```

```python
movies.shape
```

```python
movies_merged.shape
```

```python
movies_merged.head(2)
```

```python
movies_merged_cleaned = movies_merged.drop(columns=['homepage','status','title_x','title_y','production_countries'])
```

```python
movies_merged_cleaned.head(3)
```

```python
movies_merged_cleaned.shape
```

```python
movies_merged.shape
```

```python
movies_merged_cleaned.info()
```

```python
##Using Weighted Average for each Average Rating
```

```python
# W= Weighted Rating
# R= Average for the movie as a number from 0 to 10 (mean)= (Rating)
# v= number of votes for the movie = (votes)
# C= the mean vote across the whole report (currently 6.9)
# m= minimum votes required to be listed in top 250 (Currently 3000)
```

```python
movies_merged_cleaned.shape
```

```python
# Calculate all the components based on the above formula
v=movies_merged_cleaned['vote_count']
R=movies_merged_cleaned['vote_average']
C=movies_merged_cleaned['vote_average'].mean()
m=movies_merged_cleaned['vote_count'].quantile(0.70)
```

```python
movies_merged_cleaned.shape
```

```python
movies_merged_cleaned['weighted_average']=((R*v)+ (C*m))/(v+m)
```

```python
movies_merged_cleaned.head(2)
```

```python
movies_merged_cleaned.shape
```

```python
movies_merged_cleaned.info()
```

```python
movie_m_sorted=movies_merged_cleaned.sort_values('weighted_average',ascending=False)
```

```python
movie_m_sorted.head(3)
```

```python
movie_m_sorted[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(30)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
weight_average= movie_m_sorted.sort_values('weighted_average',ascending=False)
```

```python
plt.figure(figsize=(12,6))
axis1=sb.barplot(x=weight_average['weighted_average'].head(10), y=weight_average['original_title'].head(10), data=weight_average)
plt.xlim(0,10)
plt.title('Best Movies by Average Votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
popularity= movie_m_sorted.sort_values('popularity',ascending=False)

plt.figure(figsize=(12,6))
axis1=sb.barplot(x=popularity['popularity'].head(10), y=popularity['original_title'].head(10), data=popularity)
plt.xlim(0,1000)
plt.title('Most Popular Movies by Votes', weight='bold')
plt.xlabel('Popularity Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```

```python
## Recommendation based on scaled weighted average and popularity score(Priority is given 50% to both)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
movie_scaled=scaling.fit_transform(movies_merged_cleaned[['weighted_average','popularity']])
movie_normalized=pd.DataFrame(movie_scaled,columns=['weighted_average','popularity'])
movie_normalized.head()
```

```python
movies_merged_cleaned[['normalized_weight_average', 'normalized_popularity']]= movie_normalized
```

```python
movies_merged_cleaned.head(3)
```

```python
movies_merged_cleaned.shape
```

```python
movies_merged_cleaned['score']= movies_merged_cleaned['normalized_weight_average']*0.5 + movies_merged_cleaned['normalized_popularity']*0.5
movies_scored = movies_merged_cleaned.sort_values(['score'], ascending=False)
movies_scored[['original_title','normalized_weight_average','normalized_popularity','score']].head(30)
```

```python
scored= movies_merged_cleaned.sort_values('score',ascending=False)

plt.figure(figsize=(16,6))
axis1=sb.barplot(x=scored['score'].head(10), y=scored['original_title'].head(10), data=scored)
plt.xlim(0,1)
plt.title('Best Rated & Most Popular Movies', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_movies.png')
```
